```
# @markdown Loading Dependencies
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```
# @markdown Fetching District names
states = ['bihar','rajasthan','maharashtra','madhya-pradesh','haryana']

india = {}
for i in states:  
  stateLink = 'https://www.nriol.com/india-statistics/' + i + '/districts.asp'
  print(stateLink)
  statePage = requests.get(stateLink)
  print(statePage)
  stateSoup = bs(statePage.content,'html.parser')
  tableValues = stateSoup.find_all('td')
  districts = []
  for j in range(1,len(tableValues),5):
    districts.append(tableValues[j].get_text())
  india[i] = districts
```

```
def toCelsius(temp):
  fahrenheit = int(temp)
  celsius = (fahrenheit - 32) * 5/9
  celsius = int(round(celsius,0))
  return celsius
```

```
# @markdown Dataset Scraping
latitude =[]
longitude = []
district = []
month = []
irradiance = []
state = []
temperature = []
for key,value in india.items():
  print(key)
  print(value)
  for i in value:
    link = 'http://www.synergyenviron.com/tools/solar-irradiance/india/' + key + '/' + i
    print(link)
    link2 = 'https://www.timeanddate.com/weather/india/' + i.lower() + '/climate'
    print(link2)
    try:
      page = requests.get(link)
      soup = bs(page.content,'html.parser')
      small = soup.find_all('small')
      td2 = soup.find_all('td',class_ = 'small')

      forLatLon= (small[0].get_text()).split(' ')
      la = forLatLon[2]
      lt = forLatLon[6]
      
      weatherPage = requests.get(link2)
      soup2 = bs(weatherPage.content,'html.parser')
      all_months = soup2.find_all('div',class_ = 'climate-month')
      #print(all_months)
      if all_months:
        for k in range(1,len(all_months)):
          mean_temp = (all_months[k].get_text()).split('\xa0°F')[2]
          temp = mean_temp.split(' ')[3]
          celsius = toCelsius(temp)
          temperature.append(celsius)
          #print(temperature)
        
        for j in range(12,int(len(td2)/2)):
          month.append(td2[2*j].get_text())
          #print(month)
          irradiance.append(td2[2*j+1].get_text())
          #print(irradiance)
          latitude.append(la)
          #print(latitude)
          longitude.append(lt)
          #print(longitude)
          district.append(i)
          #print(district)
          state.append(key)
    except:
      pass

    
```

```
# @markdown Dataset scraping manually for some important districts
value = ['shahdara','saket','connaught-place','kanjhawala','rajouri-garden','sadar-bazar','vasant-vihar','darya-ganj','preet-vihar']
t_delhi = [14 , 17 , 23 , 29 , 33 , 33 , 31 , 30 , 29 , 26 , 20 , 15 ]
for city in value:
    link = 'http://www.synergyenviron.com/tools/solar-irradiance/india/' + 'delhi' + '/' + city
    print(link)
    try:
      page = requests.get(link)
      soup = bs(page.content,'html.parser')
      small = soup.find_all('small')
      forLatLon= (small[0].get_text()).split(' ')
      la = forLatLon[2]
      lt = forLatLon[6]
      
      td2 = soup.find_all('td',class_ = 'small')

      for j in range(12,int(len(td2)/2)):
        month.append(td2[2*j].get_text())
        irradiance.append(td2[2*j+1].get_text())
        latitude.append(la)
        longitude.append(lt)
        district.append(city)
        state.append('delhi')
    except:
      pass

for t in range(108):
    temperature.append(t_delhi[t%12])
      
mydict = {'State':state,'District' : district,'Latitude' : latitude , 'Longitude' : longitude,'Month': month,'Irradiance' : irradiance, 'Temperature': temperature}
df = pd.DataFrame(mydict)
df.to_csv('/content/gdrive/My Drive/minor/Irradiance_Dataset2.csv') 
```
